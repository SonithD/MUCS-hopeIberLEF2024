```python
#eemot_object = emot.core.emot()
ps =PorterStemmer()
lemmatiser = WordNetLemmatizer()
english_stopwords = stopwords.words('english')
exclude = set(string.punctuation)
def preprocess(text):
  text = contractions.fix(text.lower(), slang=True)
  text =re.sub("@ ?[A-Za-z0-9_]+", " ", text)
  text = text.replace('$', ' dollar ')
  text=re.sub(r'$', '', text)
  text= re.sub(r'’','', text)
  text=re.sub(r'¿','',text)
  text=re.sub('<.*?>','',text)
  text=re.sub(r'http\S+', '', text)
  text = ''.join(ch for ch in text if ch not in exclude)
  tokens = word_tokenize(text)
  #print("Tokens:", tokens)
  text = [t for t in tokens if t not in english_stopwords]
  text = " ".join(text)
  return text
```

```python
def number_to_words(text):
    return ' '.join(p.number_to_words(token) if token.isdigit() else token for token in text.split())
```

```python
import emoji
#import demoji
#demoji.download_codes()
def emo(text):
  temp=emoji.demojize(text,delimiters=(" "," "))
  temp=temp.replace("_","  ")
  return temp
```

```python
df_train['clean_text']=df_train["text"].apply(lambda x:emo(x))
df_train["clean_text"]=df_train['clean_text'].apply(lambda X: number_to_words(X))
df_train["clean_text"]=df_train['clean_text'].apply(lambda X: preprocess(X))
```

```python
df_val['clean_text']=df_val["text"].apply(lambda x:emo(x))
df_val["clean_text"]=df_val['clean_text'].apply(lambda X: number_to_words(X))
df_val["clean_text"]=df_val['clean_text'].apply(lambda X: preprocess(X))
```

```python
# Move last Column to First Column
new_cols = ["clean_text","category"]
df_train=df_train[new_cols]
df_train
```

```python

# Import label encoder
from sklearn import preprocessing

# label_encoder object knows
# how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df_train['category']= label_encoder.fit_transform(df_train['category'])

df_train['category'].unique()
```

```python
model = ClassificationModel('bert', 'mrm8488/distill-bert-base-spanish-wwm-cased-finetuned-spa-squad2-es', num_labels=2, args={'reprocess_input_data': True, 'overwrite_output_dir': True}, use_cuda=False)
#juanmvs/homophobia-mexican-spanish
```

```python
model.train_model(df_train)
```

```python
list_val_text1 = []
for text in df_val['clean_text']:
  list_val_text1.append(text)
```

```python
val_pred, raw_val_outputs = model.predict(list_val_text1)
```

```python
predctval=label_encoder.inverse_transform(val_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(df_val['category'], predctval))
```

```python
list_test_text = []
for text in df_test['clean_text']:
  list_test_text.append(text)
```
